**3.2 Encoding Categorical Variables**


```python
print("\n 2. Encoding Categorical Variables")
print("--" *30 )
##Note this target is already numerically encoded in 'target' (0, 1)
print(f"The target variable is already numerically encoded in 'target' column")
print(f"I will keep the 'diagonis' column for reference and drop it before moel training")
print("--" *30 )

##Separate the variables
x = df.drop(['target', 'diagnosis'], axis=1)
##x = df[breast_cancer.feature_names]
y = df['target']
print(f"\nShape of features:\n\n{x.head()}")
print("--" *30 )
print(f"Shape of target:\n\n{y.head()}")
print("--" *30 )
```

**3.3 Normalizing Features**

```python
from sklearn.preprocessing import StandardScaler
#As seen from the EDA feature scaling is essential due to different scales
scaler = StandardScaler()

#Fitting the scaler and transforming the features
x_scaled = scaler.fit_transform(x)

#Converting scaled array back to a DataFrame for clear reading
x_scaled_df = pd.DataFrame(x_scaled, columns = x.columns)
#x_scaled_df = pd.DataFrame(x_scaled, columns = breast_cancer.feature_names)
print(x_scaled_df.head())
print("--" *30 )
print(f"-Features have been Standardized using StandardScaler")
print("-Mean of the first features after scaling:", f"{x_scaled_df['mean radius'].mean():.6f}")
print("-Std Dev of the first features after scaling:", f"{x_scaled_df['mean radius'].std():.6f}")
print("--" *30 )
```

**3.4 Spliting Dataset 80% for training and 20% for testing**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state= 42, stratify = y)
print("--" *30 )
print(f"Dataset split completed")
print(f"x_train shape: {x_train.shape}: y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}: y_test shape: {y_test.shape}")
print("--" *30 )

#Checking class distribution in split to make sure stratification worked
print(f"Class distribution in y_train\n{pd.Series(y_train).value_counts().sort_index()}")
print("--" *30 )
print(f"Class distribution in y_test\n{pd.Series(y_test).value_counts().sort_index()}")
print("--" *30 )
```

**4. Model Selection**

I am choosing the following models
1. Logistic Regression a linear Model
2. k-Nearest Neighbors an instance-based learning model, makes predictions based on local similarity of data points
3. Random Forest an ensemble method, a non-linear ensemble method rpbust to overfitting

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


#Initializing the models
#Here we use class_weight = 'balanced'in LogisticRegression and Random Forest to handle the imbalance
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=42, class_weight = 'balanced'),
    'k-Nearest Neighbors': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(random_state =42, class_weight='balanced')
}
```

**4.2 Training Models**

```python
print("\nTraining models and Predicting")
print("--"*20)
#Dictionary to store the results
trained_models = {}

#Training models
for name, clf in classifiers.items():
  clf.fit(x_train, y_train)
  trained_models[name] = clf
  print(f"{name} trained successfully")
print("--"*20)
print(f"All models are trained on {x_train.shape[0]} data sample")
```

**4.3 Predicting the model using Test data**


```python
results = {}

for name, clf in trained_models.items():
  #Making predictions
  y_pred = clf.predict(x_test)

  #Accuracy
  accuracy = accuracy_score(y_test, y_pred)

  #precision = precision_score(y_test, y_pred)
  #recall = recall_score(y_test, y_pred)
  #f1 = f1_score(y_test, y_pred)


  #Store prediction
  results[name] = {
      'model': clf,
      'accuracy': accuracy,
      'prediction': y_pred,
      #'precision': precision,
      #'recall': recall,
      #'f1': f1
  }
  print(f"{name}: Test Accuracy = {accuracy:.3f}")

print(f"\nModel Prediction Comparison ")
print("--"*50)
for name, result in results.items():
  print(f"{name:20s}: {result['prediction']}")
  print("--"*50)
print(f"All models are tested on {y_test.shape[0]}")


```

**ROC Curves**

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
axes = axes.ravel()
colors = ['blue', 'red', 'green']

# Helper function to plot a ROC curve
def plot_roc_curve(axis, model, x_test, y_test, color, model_name):

    y_pred_proba = model.predict_proba(x_test)[:, 1]
    # Calculating ROC Curve and AUC
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)

    axis.plot(fpr, tpr, color=color, lw=2, label=f'{model_name} (AUC = {roc_auc:.3f})')
    axis.plot([0, 1], [0, 1], 'k--', lw=2)

    axis.set_xlim([0.0, 1.0])
    axis.set_ylim([0.0, 1.05])
    axis.set_xlabel('False Positive Rate')
    axis.set_ylabel('True Positive Rate')
    axis.set_title(f'ROC Curve: {model_name}')
    axis.legend(loc="lower right")


for idx, (name, model) in enumerate(classifiers.items()):
    plot_roc_curve(axes[idx], model, x_test, y_test, colors[idx], name)

plt.tight_layout()
plt.show()
```
